## üëâ Quantity affects the Sales
### üëâ  Discount affects the Sales

```python
df.hist(figsize=(10,10),color='pink');
```

### üëâ There is higher sales and profit of Consumer and the Home Office has the least sales and profit in relation to segments

```python
df.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['pink','brown'],figsize=(8,5))
plt.ylabel('Profit and sales ')
plt.show()
```

#### üëâ Technology category has the highest sale and profit and Furniture category has least sales and proft in relation to Category of product

```python
df.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['lightblue','darkblue'],figsize=(8,5))
plt.ylabel('Profit and sales ')
plt.show()
```

## üëâ The PHONES and CHAIRS has same number of sales but Phones creates more profits than chairs***
### üëâ Fasterner has least sales with no profit
#### üëâ bookcases and tables has good sales but there is loss

```python
plt.figure(figsize=(20,10))
df.groupby('Sub-Category')[['Profit','Sales']].sum().plot.bar(color=['red','lightblue'],figsize=(8,5))
plt.ylabel('Profit and sales ')
plt.show()
```

## üëâ There is higher sales when products are shipped using Standard class ship mode
### üëâ People uses Standard Class more than any other ship mode
#### üëâ Same Day ship mode is least used by customers

```python
df.groupby('Ship Mode')[['Profit','Sales']].sum().plot.bar(color=['darkorange','lightblue'],figsize=(8,5))
plt.ylabel('Profit and sales ')
plt.show()
```
